# Описание проекта<a class="anchor" id="b1"></a>


Оператор мобильной связи «Мегалайн» выяснил: многие клиенты пользуются архивными тарифами. Они хотят построить систему, способную проанализировать поведение клиентов и предложить пользователям новый тариф: «Смарт» или «Ультра».

В распоряжении данные о поведении клиентов, которые уже перешли на эти тарифы (см. проект 5 в этом же репозитории). Нужно построить модель для задачи классификации, которая выберет подходящий тариф. Предобработка данных не требуется, так как она уже проведена.

Необходимо построить модель с максимально большим значением accuracy (>0.75). Для этого рассмотреть различные семейства моделей с различными гиперпараметрами.

## Описание данных
Каждый объект в наборе данных — это информация о поведении одного пользователя за месяц. Известно:
* сalls — количество звонков,
* minutes — суммарная длительность звонков в минутах,
* messages — количество sms-сообщений,
* mb_used — израсходованный интернет-трафик в Мб,
* is_ultra — каким тарифом пользовался в течение месяца («Ультра» — 1, «Смарт» — 0).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from time import time
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.dummy import DummyClassifier
```

# 1. Подготовка данных<a class="anchor" id="b2"></a>

Загрузим данные и посмотрим общую информацию

```python
data = pd.read_csv('data/users_behavior.csv')
data.head()
```

```python
data.info()
```

Всего 3214 объектов, пропусков нет. Целевой признак -- is_ultra. Он принимает значения 0 или 1 (таким образом, перед нами задача бинарной классификации). Посмотрим долю объектов с целевыми признаками 0 и 1.

```python
data['is_ultra'].value_counts()/len(data)
```

Объектов с целевым признаком, равным 1 -- около 31% от всей выброки, то есть значения представлнены неравномерно. 

Разделим выборку на обучающую, валидационную и тестовую в отношении 3:1:1 с учетом неравномерности представления целевых признаков.

```python
features = data.drop(['is_ultra'], axis=1)
target = data['is_ultra']

features_train, features_tv, target_train, target_tv = train_test_split(
    features, target, test_size=0.4, random_state=1, stratify=target)
features_valid, features_test, target_valid, target_test = train_test_split(
    features_tv, target_tv, test_size=0.5, random_state=1, stratify=target_tv)
print(f'Размеры выборок: обучающей: {len(features_train)}, валидационной: {len(features_valid)}, тестовой: {len(features_test)}')

for tr in [target_train, target_valid, target_test]:
    print(tr.value_counts()/len(tr))
```

# 2. Исследование качества различных моделей c различными наборами гиперпараметров<a class="anchor" id="b3"></a>

Будем сравнивать по метрике accuracy следующие модели с различными параметрами:
* DecisionTreeClassifier,
* RandomForestClassifier,
* LogisticRegression.

Сначала проведем подбор гиперпараметров для каждого вида моделей: будем обучать модели с различными гиперпараметрами на тренировочной выборке и выбирать лучший набор гиперпараметров по accuracy на валидационной выборке. 

P.S. Все выводы о "существенном" или "несущественном" влиянии гиперпараметра на качество модели относятся конкретно (только) к этой задаче и этому набору данных. В других задачах и при других выборках влияние различных гиперпараметров может сильно отличаться.

P.S.2. Времени обучения и времени предсказания всех моделей не превышает секунды, поэтому далее не рассматривается. Осталось в таблицах для подтверждения

## 2.1. DecisionTreeClassifier

Для модели DecisionTreeClassifier проварьируем следущие параметры: 

* criterion

The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “log_loss” and “entropy” both for the Shannon information gain, see Mathematical formulation.

* max_depth 

The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

* min_samples_split

The minimum number of samples required to split an internal node. If float, then min_samples_split is a fraction and ceil(min_samples_split * n_samples) are the minimum number of samples for each split.

* min_samples_leaf

The minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. This may have the effect of smoothing the model, especially in regression. If float, then min_samples_leaf is a fraction and ceil(min_samples_leaf * n_samples) are the minimum number of samples for each node.

```python
columns = ['criterion', 'max_depth', 'min_samples_split', 'min_samples_leaf', 'accuracy_score', 'trainig_time', 'predicting_time']
data = []

for criterion in ['gini', 'entropy']:
    for max_depth in range(3, 28, 4):
        for min_samples_split in [0.0001, 0.001, 0.01]:
            for min_samples_leaf in [0.0001, 0.001, 0.01]:
                model = DecisionTreeClassifier(criterion=criterion,
                        max_depth=max_depth,
                        min_samples_split=min_samples_split,
                        min_samples_leaf=min_samples_leaf,
                        random_state=0)
                
                start = time()
                model.fit(features_train, target_train)
                end = time()
                training_time = round(end - start)
                
                start = time()
                predictions = model.predict(features_valid)
                end = time()
                predicting_time = round(end - start)
                
                score = accuracy_score(predictions, target_valid)
                string = [criterion, max_depth, min_samples_split, min_samples_leaf,
                         score, training_time,predicting_time]
                data.append(string)
```

```python
tree_metrics = pd.DataFrame(columns=columns, data=data)
print("10 лучших наборов гиперпараметров")
tree_metrics.sort_values('accuracy_score', ascending = False).head(10)
```

```python
print("10 худших наборов гиперпараметров")
tree_metrics.sort_values('accuracy_score', ascending = True).head(10)
```

И в топ-10 лучших, и в топ-10 худших наборах гиперпараметров использовались оба критерия: entropy и gini, а значит критерий в данной задаче не оказывает значительного влияния. 

При измерениях accuracy на валидационной выборке наилучшей максимальной глубиной явно является глубина, равная 7: именно такое значение она принимает во всех 10 лучших наборов гиперпараметров. Увеличение глубины не привело к улучшению модели: видимо, она переобучалась на тренировочной выборке. И слишком маленькая глубина (3) не позволила получить конкурентные результаты. 

Параметры min_samples_split и min_samples_leaf в данном случае не оказывали существенного влияния: их большие и маленькие значения попали как в лучшие, так и в худшие значения.

Таким образом, можно сказать, что при решении поставленной задачи наиболее важным гиперпарамтром является максимальная глубина дерева.

Разница между худшим и лучшим результатом -- 0.1, то есть качество модели довольно сильно зависит от подбора гиперпараметров.

Сохраним лучшие наборы гиперпараметров 

```python
best_tree = tree_metrics.sort_values('accuracy_score', ascending = False).head(1)
best_tree
```

## 2.2. RandomForestClassifier

RandomForestClassifier
Для модели RandomForestClassifier проварьируем следущие параметры:

* n_estimators
The number of trees in the forest.

* class_weight

Weights associated with classes in the form {class_label: weight}. If not given, all classes are supposed to have weight one. 

* bootstrap

Whether bootstrap samples are used when building trees. If False, the whole dataset is used to build each tree.

```python
columns = ['n_estimators', 'class_weight', 'bootstrap', 'accuracy_score', 'trainig_time', 'predicting_time']
data = []

for n_estimators in range(20, 300, 20):
    for class_weight in ['balanced', None]:
        for bootstrap in [True, False]:
            # for min_samples_leaf in [0.0001, 0.001, 0.01]:
                model = RandomForestClassifier(n_estimators=n_estimators,
                        class_weight=class_weight,
                        bootstrap=bootstrap,
                        random_state=0)
                
                start = time()
                model.fit(features_train, target_train)
                end = time()
                training_time = round(end - start)
                
                start = time()
                predictions = model.predict(features_valid)
                end = time()
                predicting_time = round(end - start)
                
                score = accuracy_score(predictions, target_valid)
                string = [n_estimators, class_weight, bootstrap,
                         score, training_time,predicting_time]
                data.append(string)
```

```python
forest_metrics = pd.DataFrame(columns=columns, data=data)
print("10 лучших наборов гиперпараметров")
forest_metrics.sort_values('accuracy_score', ascending = False).head(10)
```

```python
print("10 худших наборов гиперпараметров")
forest_metrics.sort_values('accuracy_score', ascending = True).head(10)
```

Разница между худшим и лучшим результатом -- 0.01, то есть качество модели RandomForestClassifier намного меньше зависит от подбора гиперпараметров, чем качество модели DecisionTreeClassifier.

Для данной модели явно наибольшее влияние оказывает гиперпараметр bootstrap: во всех "лучших" моделях он принимает значение True, во всех худших -- False. То есть использование всего датасета для обучения каждого дерева негативно сказывается на качестве модели: теряется разнообразие ответов при голосовании.

Бесконечное увеличение количества деревьев не приводит к улучшению качества модели: она переобучется. У лучшей модели не самое большое число n_estimators. 

```python
best_forest = forest_metrics.sort_values('accuracy_score', ascending = False).head(1)
best_forest
```

## 2.3. LogisticRegression


Для модели LogisticRegression проварьируем следущие параметры: 

* penalty

Specify the norm of the penalty

* tol 

Tolerance for stopping criteria.

* solver

Algorithm to use in the optimization problem. 

* max_iter

Maximum number of iterations taken for the solvers to converge.


```python
columns = ['tol', 'max_iter', 'solver', 'penalty', 'accuracy_score', 'trainig_time', 'predicting_time']
data = []

for tol in [1e-3, 1e-4, 1e-5]:
    for max_iter in range(200, 451, 100):
        for solver in ['lbfgs', 'sag']:
            for penalty in ['l2', 'none']:
                model = LogisticRegression(tol=tol,
                        max_iter=max_iter,
                        solver=solver,
                        penalty=penalty,
                        random_state=0)
                
                start = time()
                model.fit(features_train, target_train)
                end = time()
                training_time = round(end - start)
                
                start = time()
                predictions = model.predict(features_valid)
                end = time()
                predicting_time = round(end - start)
                
                score = accuracy_score(predictions, target_valid)
                string = [tol, max_iter, solver, penalty,
                         score, training_time,predicting_time]
                data.append(string)
```

```python
regression_metrics = pd.DataFrame(columns=columns, data=data)
print("10 лучших наборов гиперпараметров")
regression_metrics.sort_values('accuracy_score', ascending = False).head(10)
```

```python
print("10 худших наборов гиперпараметров")
regression_metrics.sort_values('accuracy_score', ascending = True).head(10)
```

Разница между лучшим и худшим результатом -- 0.06, то есть для модели LogisticRegression влияние гиперпараметров на ее качество -- среднее, по сравнению с деревом и лесом.

Явно прослеживается сильное влияние одного параметра -- solver: в лучших наборах он равен lbfgs, в худших -- sag. В документации сказано, что алгоритм sag подходит для больших датасетов, так как он быстрее сходится. У нас же значений немного (в обучающей выборке около 2 тысяч, поэтому здесь лучше подошел другой solver.

Влияние других гиперпараметров на качество модели слабее и более разбросано.

```python
best_regression = regression_metrics.sort_values('accuracy_score', ascending = False).head(1)
best_regression
```

Проверим качество данной модели с данными гиперпараметрами на тестовой выборке.

```python
model = DecisionTreeClassifier(criterion='entropy',
        max_depth=7,
        min_samples_split=0.0001,
        min_samples_leaf=0.0001,
        random_state=0)

model.fit(features_train, target_train)
score = model.score(features_test, target_test)
print('accuracy_score на тестовой выборке:', score)
```

accuracy_score на тестовой выборке оказался даже немного выше, чем на валидационной. Требование, указанное в задании (accuracy_score > 0.75) выполнено. 

# 4. Проверка модели на адекватность<a class="anchor" id="b5"></a>

```python
print("accuracy_score на тестовых данных:", score)

dummy_clf = DummyClassifier(strategy="uniform", random_state = 12)
dummy_clf.fit(features_train, target_train)
print("accuracy_score на рандомных 0 и 1:", accuracy_score(dummy_clf.predict(features_test), target_test))

dummy_clf = DummyClassifier(strategy="stratified", random_state = 12)
dummy_clf.fit(features_train, target_train)
print("accuracy_score на стратифицированной выборке:", accuracy_score(dummy_clf.predict(features_test), target_test))

dummy_clf = DummyClassifier(strategy="most_frequent", random_state = 12)
dummy_clf.fit(features_train, target_train)
print("accuracy_score на наиболее частых значениях:", accuracy_score(dummy_clf.predict(features_test), target_test))

dummy_clf = DummyClassifier(strategy="constant", random_state = 12, constant = 0)
dummy_clf.fit(features_train, target_train)
print("accuracy_score на нулях:", accuracy_score(dummy_clf.predict(features_test), target_test))

dummy_clf = DummyClassifier(strategy="constant", random_state = 12, constant = 1)
dummy_clf.fit(features_train, target_train)
print("accuracy_score на единицах:", accuracy_score(dummy_clf.predict(features_test), target_test))
```

Обученная модель показывает себя лучше, чем все рассмотренные примтивные "предсказатели", а значит ее можно назвать адекватной.
