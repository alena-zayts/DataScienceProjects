Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Необходимо построить модель с предельно большим значением F1-меры (>=0.59). Дополнительно измерять AUC-ROC, сравнивать её значение с F1-мерой.

Источник данных: https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling

Данные находятся в файле /data/Churn.csv.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import (
    precision_score, recall_score, f1_score, classification_report, 
    precision_recall_curve, roc_curve, roc_auc_score
)
````

# 1. Загрузка и подготовка данных<a class="anchor" id="b2"></a>

````python
data = pd.read_csv('data/Churn.csv')
data.head()
````

Столбецы RowNumber, CustomerId, Surname не несут полезной информации, избавимся от них. 

````python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
data.info()
````

ВСего 10000 записей. В столбце Tenure (количество недвижимости у клиента) есть около 1000 пропусков. Посмотрим, какие значения в этом столбце представлены в принципе. 

````python
data['Tenure'].value_counts()
````

Можно было бы предположить, что это означает, что у пользователя нет недвижимости, и заполнить пропуски нулями, однако нули встречаются как отдельные значения, поэтому такая обработка пропусков, скорее всего, будет некорректной. Посмотрим на строки с пустыми значениями в Tenure: возможно, существует некая закономерность.

````python
data[data['Tenure'].isnull()]
````

Явной закономерности не прослежтивается. Остаются варианты -- запонить пропуски медианой/средним по столбцу или же в принципе отбросить строки с пропусками. Таких строк около 10% -- это достаточно много. Однако заполнение медианой/средним может не соответсвовать действительности и сильно исказить реальные данные. Поэтому принимается решение отбросить строки с пропусками.

````python
data = data.dropna()
data.info()
data.head()
````

Осталась выборка из 9091 записей. Выполним преобразование типов: в Tenure хранятся только целочисленные значения (видно в выводах выше), однако тип данных указан как float. Преобразуем этот столбец к int. Также к целочисленным значениям преобразуем столбцы Balance и EstimatedSalary.

````python
for column in ['Tenure', 'Balance', 'EstimatedSalary']:
    data[column] = data[column].astype(int)
data
````

**Перед нами задача бинарной классификации**: уйдёт клиент из банка в ближайшее время или нет (exited=1 или 0).


Подготовим категориальные признаки (столбцы geography и gender): преобразуем их в численные техникой прямого кодирования (One-Hot Encoding, OHE), так как она подходит для решения нашей задачи любой моделью (в отличие, например, от техники порядкового кодирования (Ordinal Encoding), которая не подходит для модели логистической регрессии). 

Чтобы не попасть в дамми-ловушку (dummy trap, ловушка фиктивных признаков/излишней информации), в функции pd.get_dummies() укажем аргумент drop_first=True, чтобы отбросить один столбец, который легко восстанавливается по остальным.  

````python
data = pd.get_dummies(data, drop_first=True)
data
````

Количество столбцов увеличилось всего на 1 (с 11 до 12). Поле Geography принимало только 3 значения, пол -- 2.

Посмотрим, насколько сблансированной является выборка: посчитаем долю объектов каждого класса


````python
data['Exited'].value_counts() / len(data)
````

**Выборка несбалансирована**: доля объектов положительного класса (exited=1) -- лишь около 20%. Это необходимо учитывать при разделении выборки и обучении модели.

Выделим целевой признак и разделим выборку на тренировочную, валидационную и тестовую в отношении 3:1:1 с учетом дисбаланса классов

````python
target = data['Exited']
features = data.drop(['Exited'], axis=1)

features_train, features_vt, target_train, target_vt = train_test_split(
    features, target, test_size=0.4, random_state=0, stratify=target)
features_valid, features_test, target_valid, target_test = train_test_split(
    features_vt, target_vt, test_size=0.5, random_state=0, stratify=target_vt)

print(f'Размеры выборок: тренировочной {len(target_train)}, валидационной {len(target_valid)}, тестовой {len(target_test)}')

print('Доли классов сохранены:')
for df in [target_train, target_valid, target_test]:
    print(df.value_counts()/len(df))
````

Взглянем еще раз на признаки на примере тренировочной выборки.

````python
features_train
````

В признаках есть следующие численные столбцы: ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']. Посмотрим на разброс значений в этих столцах. Теперь отмасштабируем численные признаки.

````python
numeric_columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
features_train[numeric_columns].describe()
````

Можно предположить, что все эти признаки примерно одинаково важны для целевого признака, однако разброс значений в различных столбцах сильно отличается. Например, значения в столбце NumOfProducts изменяется от 1 до 4, тогда как значения в столбце Balance -- от 0 до 250898. Это может привести к тому, что модель будет считать признак Balance важнее (более сильно влияющим на ответ), чем столбец NumOfProducts. Чтобы этого избежать, отмасштабируем численные признаки методом стандартизации: среднее в каждом столбце станет равным 0, а дисперсия 1.

````python
pd.options.mode.chained_assignment = None 
scaler = StandardScaler()
scaler.fit(features_train[numeric_columns])

features_train[numeric_columns] = scaler.transform(features_train[numeric_columns])
features_valid[numeric_columns] = scaler.transform(features_valid[numeric_columns])
features_test[numeric_columns] = scaler.transform(features_test[numeric_columns])
features_train
````

Данные готовы для обучения моделей и их исследования: пропуски обработаны, категориальные признаки преобразованы к численным, а численные признаки стандартизированы. Выделен целевой признак, исходная выборка разделена на обучающую, валидационную и тестовую.

# 2. Обучение моделей без учета дисбаланса классов<a class="anchor" id="b3"></a>

**В данной работе будем пробовать обучать три семейства моделей с различными гиперпараметрами, оценивая их метрикой F1: логистическую регрессию (LogisticRegression), решающее дерево (DecisionTreeClassifier) и лес (RandomForestClassifier).**

Для 3 лучших моделей (модели с лучшим набором гиперпараметров в каждом классе) по валидационной выборке будем дополнительно считать roc_auc_score, precision, recall, а также строить кривые precision-recall и Receiver Operating Characteristic (ROC).

````python
def find_best_models(model_grid_list, features_train, target_train):
    answer = dict()
    for name in model_grid_list:
        model, param_grid = model_grid_list[name].values()
        
        searcher = GridSearchCV(model, param_grid, scoring='f1')
        searcher.fit(features_train, target_train)
        answer[name] = {
            'best_estimator_': searcher.best_estimator_,
            'best_params_': searcher.best_params_,
            'best_score_': searcher.best_score_,
        }
    return answer

def investigate_best_models(best_models, features_valid, target_valid):
    cmp_data = []
    cmp_index = []
    
    for model_name in best_models:

        model = best_models[model_name]['best_estimator_']
        predictions = model.predict(features_valid)
        probabilities_one = model.predict_proba(features_valid)[:, 1]
        
        model_scores = dict()
        model_scores['best_params_'] = best_models[model_name]['best_params_']
        model_scores['roc_auc_score'] = roc_auc_score(target_valid, probabilities_one)
        model_scores.update(classification_report(target_valid, predictions, output_dict=True)['1'])
        cmp_data.append(model_scores)
        cmp_index.append(model_name)
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
        fig.suptitle(model)
        
        precision, recall, thresholds = precision_recall_curve(target_valid, probabilities_one)
        
        ax1.set_xlabel('Recall')
        ax1.set_ylabel('Precision')
        ax1.set_title('Precision-Recall curve')
        ax1.step(recall, precision, where='post')
        ax1.grid()
        
        fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one)
        ax2.set_xlabel('FPR')
        ax2.set_ylabel('TPR')
        ax2.set_title('ROC curve')
        ax2.plot(fpr, tpr)
        ax2.grid()
        
        plt.show()
        
    return pd.DataFrame(cmp_data, index=cmp_index).sort_values(by='f1-score')
````

В исследовательских целях попробуем обучить модели без учета дисбаланса классов, чтобы затем можно было оценить, насколько различные методы борьбы с дисбалансом способны улучшить качество моделей.

````python
unbalanced_model_grid_list = {
    'LogisticRegression': {
        'model': LogisticRegression(), 
        'param_grid': {
            'tol': [1e-7, 1e-6, 1e-5, 1e-4,], 
            'max_iter': range(50, 201, 50),
            'penalty': ['l2', 'none']
        },
    },
    
    'DecisionTreeClassifier': {
        'model': DecisionTreeClassifier(), 
        'param_grid': {
            'criterion': ['gini', 'entropy'],
            'max_depth': range(3, 16, 4),
            'min_samples_leaf': [ 0.001, 0.01, 0.1],
            
        },
    },
    
    'RandomForestClassifier': {
        'model': RandomForestClassifier(), 
        'param_grid': {
            'n_estimators': range(20, 101, 20),
            'bootstrap': [True, False],
            'max_depth': range(3, 20, 4),   
        },
    },
}

unbalanced_best_models = find_best_models(unbalanced_model_grid_list, features_train, target_train)
unbalanced_cmp_table = investigate_best_models(unbalanced_best_models, features_valid, target_valid)
````

````python
pd.set_option('max_colwidth', None)
print('Качество лучших моделей при обучении на несбалансированной выборке')
unbalanced_cmp_table
````

Таким образом, лучшее значение метрикики F1, которого удалось добиться без борьбы с дисбалансом классов, если округлять до сотых, составляет 0.57. Такой результат удалось получить с помощью моделей RandomForestClassifier с параметрами {'bootstrap': True, 'max_depth': 15, 'n_estimators': 80} и DecisionTreeClassifier с параметрами {'criterion': 'entropy', 'max_depth': 7, 'min_samples_leaf': 0.001}. RandomForestClassifier на 5 тысячных лучше.

F1 -- это среднее геометрическое precision и recall. precision больше у RandomForestClassifier, а recall у обеих моделей одинаковый. То есть модели смогли выявить одинаковое количество объектов положительного класса, и при этом RandomForestClassifier ошибался, называя отрицательные объекты положительными, реже. 

По метрике roc_auc_score RandomForestClassifier также превзошел DecisionTreeClassifier (0.84 протип 0.83), и roc-кривая выглядит более гладкой.

В целом все модели по метрике roc_auc прошли порог адекватности: их показатели выше, чем у случайной модели (0.5). Но ни одна не прошла порог по F1 метрике, который был установлен в задании (0.59)

При этом у всех моделей recall ниже, чем precision, то есть все модели стараются реже относить объекты к положительному классу, и f1 снижается в основном за счет recall. Это как раз и объясняется тем, что в выборке около 20% объектов относятся к положительному классу, и 80% -- к отрицательному (то есть отрицательных объектов в 4 раза больше, чем положительных), и модели более склонны относить объекты к отрицательному классу, так как это в целом более вероятно.

Попробуем улучшить качество моделей, сбалансировав классы.

## 3.2 Upsampling и downsampling

Теперь попробуем внести изменения в тренировочную выброку. 

1. Upsampling: увеличить выборку за счет повторения строк с положительным целевым признаком repeat=4 раза (так как отрицательных объектов примерно в 4 раза больше, чем положительных).

2. Downsampling: уменьшить выборка за счет удаления части строк с отрицательным целевым признаком (оставим из них только 25% по той же причине).


При этом необходимо перемешивать строки, чтобы не было подряд идущих схожих записей. 

Обучим на полученных выборках все модели с теми же гиперпараметрами, что и в случае, где не учитывался дисбаланс классов, выберем лучшие. Посмотрим, как изменятся f1 метрика по сравнению с обучением без учета дисбаланса для каждого семейства моделей.

````python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled


def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=0)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=0)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=0)
    
    return features_downsampled, target_downsampled
````

````python
features_train_upsampled, target_train_upsampled = upsample(features_train, target_train, 4)
upsampled_best_models = find_best_models(unbalanced_model_grid_list, features_train_upsampled, target_train_upsampled)
upsampled_cmp_table = investigate_best_models(upsampled_best_models, features_valid, target_valid)
````

````python
features_train_downsampled, target_train_downsampled = downsample(features_train, target_train, 1/4)
downsampled_best_models = find_best_models(unbalanced_model_grid_list, features_train_downsampled, target_train_downsampled)
downsampled_cmp_table = investigate_best_models(downsampled_best_models, features_valid, target_valid)
````

````python
print('(Для сравнения) При обучении на несбалансированной выборке')
unbalanced_cmp_table
````

````python
print('При обучении на сбалансированной техникой upsampling выборке')
upsampled_cmp_table
````

````python
print('При обучении на сбалансированной техникой downsampling выборке')
downsampled_cmp_table
````

````python
cmp_techniques = pd.DataFrame()
for column_name in ['f1-score', 'recall', 'precision', 'roc_auc_score']:
    cmp_techniques[f'{column_name}_upsampled'] = upsampled_cmp_table[column_name] - unbalanced_cmp_table[column_name]
    cmp_techniques[f'{column_name}_downsampled'] = downsampled_cmp_table[column_name] - unbalanced_cmp_table[column_name]

print('Изменения в качестве моделей при применении техник upsampling и downsampling в',
      'сравнении с классификацией без учета дисбаланса классов')
cmp_techniques
````

При применении каждой из техник заметный рост f1-score можно отметить только у LogisticRegression, в случае DecisionTreeClassifier он даже уменьшился в обоих случаях. roc_auc_score как уменьшиался, так и увеличивался, но в основном незначительно, иаксимальные изменения: -0.05 в DecisionTreeClassifier при upsampling и +0.02 для RandomForestClassifier при downsampling)

При этом для обеих техник для каждого семейства моделей, как и ожидалось, выросла полнота, но уменьшилась точность. 

Можно сделать вывод, что такой подход к борьбе с дисбалансом неплохо подходит для LogisticRegression, но не для DecisionTreeClassifier или RandomForestClassifier.


# 4. Выбор модели и финальное тестирование<a class="anchor" id="b5"></a>

Посмотрим на все лучшие по f1-score модели, которые удалось получить на данный момент.

````python
pd.concat(
    [unbalanced_cmp_table, class_weight_cmp_table, upsampled_cmp_table, downsampled_cmp_table],  
    keys=['unbalanced', 'class_weight', 'upsampled', 'downsampled']
).sort_values(by='f1-score', ascending=False)
````

Можно сделать вывод, что модель, обученная с учетом дисбаланса, справляется с прогнозами лучше, чем без его учета, так как получает более сбалансированную выборку и уже не склонна всегда отдавать предпочтение более частому классу.

Обучим лучшую модель на тренировочной выборке и проведем финальное тестирование на тестовой.

````python
model = RandomForestClassifier(random_state=0, bootstrap=True, class_weight='balanced', max_depth=7, n_estimators=100)
model.fit(features_train, target_train)
predictions = model.predict(features_test)
test_probabilities= model.predict_proba(features_test)
test_probabilities_one = test_probabilities[:, 1]

scores = classification_report(target_test, predictions, output_dict=True)['1']
scores.update({'roc_auc': roc_auc_score(target_test, test_probabilities_one)})
print('Результаты финального тестирования выбранной модели')
for score in scores.items():
    print(score)
````

Удалось перейти порог F1 метрики: ее значение на тестовой выборке 0.62 (требовалось >=0.59). 

При этом полнота лучше точности (0.74 против 0.53), то есть модель довольно неплохо находит объекты положительного класса, но при этом и часто ошибается, называя отрицательные объекты положительными. Устраивают ли такие значения точности и полноты -- зависит от требований заказчика.

Значение roc_auc также довольно высокое: 0.87 (в сравнении со случайной моделью 0.5)
